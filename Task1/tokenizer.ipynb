```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual file name)
df = pd.read_csv('100000_captioned_dataset.csv')

df.head()
```

```python
from transformers import BertTokenizer

# Load pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Replace NaN values with an empty string before tokenization
df['content'] = df['content'].fillna('')
df['img_caption'] = df['img_caption'].fillna('')

# Tokenize "content" column
df['content_tokens'] = df['content'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, padding='max_length', max_length=512))

# Tokenize "img_caption" column
df['img_caption_tokens'] = df['img_caption'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, padding='max_length', max_length=512))

# Show the tokenized columns
df[['content', 'content_tokens', 'img_caption', 'img_caption_tokens']].head()
```

```python
# Save the tokenized dataset to CSV
df[['content', 'content_tokens', 'img_caption', 'img_caption_tokens']].to_csv('tokenized_dataset.csv', index=False)

# Download the file
from google.colab import files
files.download('tokenized_dataset.csv')
```
